```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
data = np.random.randn(1000)
```

```python
fig = plt.figure(figsize=(14, 10))
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 2, 3)
ax4 = plt.subplot(2, 2, 4, projection='3d')
```

```python
ax1.plot(x, y1, label='sin(x)', color='blue')
ax1.plot(x, y2, label='cos(x)', color='red', linestyle='--')
ax1.legend()
ax1.set_title('Trigonometric Functions')
ax1.grid()
```

```python
ax2.hist(data, bins=30, color='purple')
ax2.set_title('Histogram')
```

```python
x_s = np.random.rand(100)
y_s = np.random.rand(100)
c = np.random.rand(100)
sc = ax3.scatter(x_s, y_s, c=c, cmap='viridis')
plt.colorbar(sc, ax=ax3)
```

```python
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax4.plot_surface(X, Y, Z, cmap='coolwarm')
```

```python
plt.tight_layout()
plt.show()
```
